```python
# -------------------------------
# Step 1: Import Libraries
# -------------------------------
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# -------------------------------
# Step 2: List All Uploaded CSV Files
# -------------------------------
folder_files = os.listdir()
csv_files = [f for f in folder_files if f.endswith('.csv')]

print("All CSV files:", csv_files)

# -------------------------------
# Step 3: Identify Normal and Attack Files
# -------------------------------
normal_file = 'Benign_test.pcap.csv'
attack_files = [f for f in csv_files if f != normal_file]

print("Normal file:", normal_file)
print("Attack files:", attack_files)

# -------------------------------
# Step 4: Load Normal Dataset
# -------------------------------
normal_df = pd.read_csv(normal_file)
normal_df['Label'] = 0   # Normal = 0

# -------------------------------
# Step 5: Load Attack Datasets
# -------------------------------
attack_dfs = []

for file in attack_files:
    df_temp = pd.read_csv(file)
    df_temp['Label'] = 1   # Attack = 1
    attack_dfs.append(df_temp)

# Combine all attack datasets
attack_df = pd.concat(attack_dfs, ignore_index=True)

# -------------------------------
# Step 6: Combine Normal + Attack
# -------------------------------
df_all = pd.concat([normal_df, attack_df], ignore_index=True)

# Shuffle the combined dataset
df_all = df_all.sample(frac=1).reset_index(drop=True)

print("Combined dataset shape:", df_all.shape)

# -------------------------------
# Step 7: Features & Labels
# -------------------------------
X = df_all.drop('Label', axis=1)
y = df_all['Label']

# -------------------------------
# Step 8: Scaling
# -------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------
# Step 9: Train-Test Split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# -------------------------------
# Step 10: Create Federated Learning Clients
# -------------------------------
def create_clients(X, y, num_clients=3):
    client_data = []
    size = len(X) // num_clients
    for i in range(num_clients):
        start = i * size
        end = (i + 1) * size
        client_data.append((X[start:end], y[start:end]))
    return client_data

clients = create_clients(X_train, y_train, num_clients=3)

for i, client in enumerate(clients):
    print(f"Client {i+1} data shape:", client[0].shape)
```

```python
# -------------------------------
# Step 1: Import Libraries
# -------------------------------
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score

# -------------------------------
# Step 2: Define Lightweight NN
# -------------------------------
def create_model(input_dim):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_dim,)),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')  # Binary classification
    ])
    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# -------------------------------
# Step 3: Federated Learning Simulation
# -------------------------------
def federated_training(clients_data, global_epochs=5, local_epochs=1):
    input_dim = clients_data[0][0].shape[1]
    global_model = create_model(input_dim)

    for epoch in range(global_epochs):
        print(f"\n--- Global Epoch {epoch+1} ---")
        client_weights = []

        for i, (X_c, y_c) in enumerate(clients_data):
            print(f"Training Client {i+1} locally...")
            # Create a local model with global weights
            local_model = create_model(input_dim)
            local_model.set_weights(global_model.get_weights())

            # Train locally
            local_model.fit(X_c, y_c, epochs=local_epochs, batch_size=64, verbose=0)

            # Collect weights
            client_weights.append(local_model.get_weights())

        # Aggregate weights (average)
        new_weights = []
        for weights_tuple in zip(*client_weights):
            new_weights.append(np.mean(weights_tuple, axis=0))

        # Update global model
        global_model.set_weights(new_weights)

    return global_model

# -------------------------------
# Step 4: Train FL Model
# -------------------------------
global_model = federated_training(clients, global_epochs=5, local_epochs=1)

# -------------------------------
# Step 5: Evaluate on Test Set
# -------------------------------
y_pred = (global_model.predict(X_test) > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
print("\n--- Final Test Accuracy ---")
print("Accuracy:", accuracy)
```
